Con los datos del archivo "asaltos_yucatan.csv", mostrar en un mapa la ubicación de los delitos en Mérida, de ser posible muestre como _popup_ la categoría.

    
</div>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
#Cargar el archivo datos/asaltos_yucatan.csv en un DataFrame llamado data
# si es necesario usar la opcion encoding = "ISO-8859-1"
data = pd.read_csv('datos/asaltos_yucatan.csv', encoding = "ISO-8859-1")
```

```python
#Mostrar las primeras filas del DataFrame data
data.head()
```

```python
data['Categoria'].unique()
```

```python
data = data[data['Ciudad'] == 'Mérida']
```

```python
# Creando el mapa
mapa = folium.Map(location=[20.9785, -89.6332],zoom_start=12)
mapa
```

```python
mapa = folium.Map(location=[20.9785, -89.6332],zoom_start=11)
for i in range(len(data)):
    delito = data.iloc[i]
    color = 'blue'
    if delito['Categoria'] == 'asalto':
        color = 'red'
    folium.Marker(location=[delito['lat'], delito['long']],popup=f'<h3 style="color:{color};">' + delito['Categoria'] + '</h3>',
              tooltip='<strong>Click here to see Popup</strong>', icon=folium.Icon(color=color)).add_to(mapa)

mapa
```
