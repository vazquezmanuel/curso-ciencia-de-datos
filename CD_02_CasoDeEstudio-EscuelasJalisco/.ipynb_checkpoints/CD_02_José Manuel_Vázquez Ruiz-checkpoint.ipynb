Sin embargo, la diferencia no es mucha, y se puede decir que la población estudiantil ronda el 50% de hombres y 50% de mujeres, aproximadamente; esta diferencia se fue haciendo más chica conforme avanza el nivel educativo (a partir de primaria, ya que se observa una menor diferencia en el nivel preescolar).

```python
# Número promedio de docentes de EDUCACION FISICA, ACTIVIDADES ARTISTICAS, ACTIVIDADES TECNONOLOGICAS e IDIOMAS en las primarias
contador_PRIMARIA=0

contador_EDUCACION_FISICA=0
contador_ACTIVIDADES_ARTISTICAS=0
contador_ACTIVIDADES_TECNONOLOGICA=0
contador_IDIOMAS=0

with open('datos/cct_estadistica_short.csv', encoding='ISO-8859-1') as archivo_csv:
    datos_csv = list(csv.reader(archivo_csv))
    for escuela in datos_csv:
        if escuela[10] == 'PRIMARIA':
            contador_PRIMARIA += 1
            contador_EDUCACION_FISICA += int(escuela[35])
            contador_ACTIVIDADES_ARTISTICAS += int(escuela[36])
            contador_ACTIVIDADES_TECNONOLOGICA += int(escuela[37])
            contador_IDIOMAS += int(escuela[38])

print('Número promedio de docentes en las primarias (por habilidad)')
print('EDUCACION FISICA:', round(contador_EDUCACION_FISICA/contador_PRIMARIA, 2))
print('ACTIVIDADES ARTISTICAS:', round(contador_ACTIVIDADES_ARTISTICAS/contador_PRIMARIA, 2))
print('ACTIVIDADES TECNONOLOGICAS:', round(contador_ACTIVIDADES_TECNONOLOGICA/contador_PRIMARIA, 2))
print('IDIOMAS:', round(contador_IDIOMAS/contador_PRIMARIA, 2))
```

El análisis señala que, de las 4 areas, siempre hay menos de un docente en cada escuela; la interpretación que se le da quiere decir que muchas escuelas no tiene docentes especializados en las 4 areas señaladas, por lo que posiblemente sean habilidades que no están siendo desarrolladas en la educación primaria.
La habilidad menos fomentada, bajo este analisis, son las actividades tecnologicas.

```python
# Hecho con lo visto hasta ahora
lista_municipios = [] # Estructura de cada item: [municipio (str), número de escuelas (int)]
municipios_encontrados = [] # Municipios ya encontrados (str)
valores_mas_grandes = [0 for i in range(5)] # Los 5 valores más grandes encontrados en orden descendente (eg. [5, 4, 3, 2, 1])

with open('datos/cct_estadistica_short.csv', encoding='ISO-8859-1') as archivo_csv:
    next(archivo_csv)
    datos_csv = list(csv.reader(archivo_csv))
    for escuela in datos_csv:
        if escuela[8] in municipios_encontrados: # Si el municipio de la escuela ya está registrada anteriormente,
            for municipio in lista_municipios:   # entonces se itera sobre los municipios para buscar el que coincide y sumarle una unidad;
                if municipio[0] == escuela[8]:
                    municipio[1] += 1
                    break
        else:                                    # si no, poner el municipio en municipios_encontrados e inicializar lista_municipios con una unidad
            municipios_encontrados.append(escuela[8])
            lista_municipios.append([escuela[8], 1])
```

```python
# Comprobando el valor de los municipios con más escuelas
for municipio in lista_municipios:
    if municipio[1] >= valores_mas_grandes[0]:          # Si el valor del municipio actual es mayor que el más grande de la lista,
        valores_mas_grandes[4] = valores_mas_grandes[3] # bajar los demás un nivel
        valores_mas_grandes[3] = valores_mas_grandes[2]
        valores_mas_grandes[2] = valores_mas_grandes[1]
        valores_mas_grandes[1] = valores_mas_grandes[0]
        valores_mas_grandes[0] = municipio[1]           # y remplazar el valor del municipio actual como el más grande
    elif municipio[1] >= valores_mas_grandes[1]:        # Hacer lo mismo con el segundo más grande
        valores_mas_grandes[4] = valores_mas_grandes[3]
        valores_mas_grandes[3] = valores_mas_grandes[2]
        valores_mas_grandes[2] = valores_mas_grandes[1]
        valores_mas_grandes[1] = municipio[1]
    elif municipio[1] >= valores_mas_grandes[2]:        # y con el tercero, y así sucesivamente ...
        valores_mas_grandes[4] = valores_mas_grandes[3]
        valores_mas_grandes[3] = valores_mas_grandes[2]
        valores_mas_grandes[2] = municipio[1]
    elif municipio[1] >= valores_mas_grandes[3]:
        valores_mas_grandes[4] = valores_mas_grandes[3]
        valores_mas_grandes[3] = municipio[1]
    elif municipio[1] >= valores_mas_grandes[4]:
        valores_mas_grandes[4] = municipio[1]
```

```python
# Imprimir los 5 valores más grandes
for valor in valores_mas_grandes:
    for municipio in lista_municipios:
        if valor == municipio[1]:
            print(municipio[0] + ':', municipio[1])
            break
```

```python
# Utilizando diccionarios
municipios = {}

with open('datos/cct_estadistica_short.csv', encoding='ISO-8859-1') as archivo_csv:
    next(archivo_csv)
    datos_csv = list(csv.reader(archivo_csv))
    for escuela in datos_csv:
        municipios[escuela[8]] = municipios.get(escuela[8], 0) + 1

# Sort en el diccionario
municipios = sorted(municipios.items(), key=lambda x:x[1], reverse=True)
print('Los municipios que más escuelas tienen son:')
for municipio, numero in municipios[0:5]:
    print(municipio + ':', numero)
```
