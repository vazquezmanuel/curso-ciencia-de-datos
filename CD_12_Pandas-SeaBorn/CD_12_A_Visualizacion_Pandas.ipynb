# Análisis exploratorio de datos (EDA)

El análisis exploratorio de datos o Exploratory Data Analysis(EDA) es un enfoque para analizar conjuntos de datos para resumir sus principales características. Se utiliza para comprender los datos, obtener un contexto al respecto, comprender las variables y las relaciones entre ellas y formular hipótesis que podrían ser útiles a la hora de construir modelos predictivos.

Al visualizar de forma eficaz las variables de un conjunto de datos y sus relaciones,  un científico de datos puede comprender rápidamente tendencias, valores atípicos y patrones. Esta comprensión se puede utilizar para contar una historia, impulsar decisiones y crear modelos predictivos.

**En esta lección, aprenderemos cómo realizar EDA utilizando la visualización de datos. Específicamente, nos centraremos en funciones asociadas a Pandas para graficar.**


#  2: Exploración grafica con Pandas


<div class="alert alert-success">
Mostrar con una grafica de barras el numero total de datos por variedad
</div>

```python
iris['variety'].value_counts().plot(kind='bar')
```

<div class="alert alert-success">

Mostrar con graficas de líneas los valores para de las variables numéricas
</div>

```python
#Todas en una sola grafica
iris.plot(figsize = (10, 5))
```

```python
#Separadas
iris.plot(subplots=True,sharex = False,figsize = (10, 5))
plt.tight_layout()
```

```python
#Separadas en un arreglo de 2 x 2
iris.plot(subplots=True, sharex = False, sharey = False,layout = (3, 3), figsize = (8, 8))
```

<div class="alert alert-success">
Mostrar con histogramas los valores de las variables numéricas 
</div>

```python
#Histogramas separados de todas las variables númericas 
iris.hist(figsize = (8, 8))
```

```python
#Histogramas juntos de todas las variables númericas 
iris.plot.hist()
```

```python
#Histogramas juntos de todas las variables númericas con algunas modificaciones para mejorar su visualizacion
iris.plot(kind = 'hist', bins = 30, alpha=0.65)
```

```python
#Hitograma de los valores de petal_length
iris['petal_length'].plot(color = 'r',kind = 'hist', bins = 100)
```

```python
#Hitograma de los valores de petal_width y  petal_length
iris[['petal_width','petal_length',]].plot(kind = 'hist', bins = 30, alpha=0.65)
```

```python
#Histogramas juntos de todas las variables númericas pero de la varidad Virginica
iris[iris['variety']=='Virginica'].plot(kind = 'hist', bins = 50, alpha=0.65, figsize = (10, 5))
```

<div class="alert alert-success">
Mostrar con diagramas de caja y bigotes los valores de las variables numéricas 
</div>

```python
iris.plot(kind = 'box', figsize = (10, 5))
```

<div class="alert alert-success">
Mostrar un diagrama de dispersión los valores de sepal_width y petal_width 
</div>

```python
iris.plot.scatter(x = 'sepal_width', y = 'petal_width', figsize = (10, 5))
```

```python
#Se puede agregar un tercer parametro utilizando una escala de color
iris.plot.scatter(x = 'sepal_width', y = 'petal_width',
                  c = 'sepal_length', colormap='viridis', figsize = (10, 5))
```
