<div class="alert alert-success">
    
2. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para Yucatán para los meses: abril, mayo, junio, julio y agosto 
    
</div> 

```python
df_aux = df[df['Estado'] == 'Yucatán'][['Mes Acontecimiento', 'Total Hectareas']].groupby('Mes Acontecimiento').sum()
df_aux
```

```python
# Creando un dataframe con los mese solicitados
months = ['Abril', 'Mayo', 'Junio', 'Julio', 'Agosto']
data = pd.DataFrame(index=months, columns=['Hectareas'])
for month in months:
    try:
        data.loc[month] = df_aux.loc[month.lower()][0] # El cero te da solo el valor numerico
    except KeyError:
        data.loc[month] = 0
data
```

```python
# Estilo MATLAB

plt.figure(figsize=(12,6))
plt.bar(data.index, data['Hectareas'], color = 'g')

# Etiqueta de los ejes
plt.xlabel('Mes')
plt.ylabel('Total de hectareas efectadas en Yucatán')

#Titulo de la figura
plt.title('Total de hectareas afectadas en Yucatán')

plt.show()
```

```python
# Estilo orientado a objetos

# Crear una figura(fig) y el espacio (ax) para graficar
fig, ax = plt.subplots(figsize=(12,6), dpi=100)

# Crear la grafica
ax.bar(data.index, data['Hectareas'], color = 'g')

# Opciones de visualización
ax.set(title='Total de hectareas afectadas en Yucatán', xlabel='Mes',
       ylabel='Total de hectareas afectadas en Yucatán')

plt.show()
```

# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

```python
df = pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201129.csv")
df.head()
```

```python
df = df.loc[:,'nombre':].set_index('nombre').T
df.head()
```

```python
df = df.reset_index()
df['index'] = pd.to_datetime(df['index'], format = '%d-%m-%Y')
df = df.set_index('index')
df.head()
```

<div class="alert alert-success">
    
1. Hacer una grafica de barras comparando todos los meses, considerando el total nacional (por mes) 
    
</div> 

```python
data = df['Nacional'].groupby(df.index.month).sum()
data
```

```python
months = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
          'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre']
```

```python
# Estilo MATLAB

plt.figure(figsize=(12,6))
plt.bar(data.index, data.values, color='green')

# Etiqueta de los ejes
plt.xlabel('Mes')
plt.xticks(np.arange(11) + 1, months)
plt.ylabel('Casos confirmados')

#Titulo de la figura
plt.title('Casos confirmados de covid, por mes')

plt.show()
```

```python
# Estilo orientado a objetos

# Crear una figura(fig) y el espacio (ax) para graficar
fig, ax = plt.subplots(figsize=(12,6), dpi=100)

# Crear la grafica
ax.bar(data.index, data.values, color='green')

# Opciones de visualización
ax.set(title='Casos confirmados de covid, por mes', xlabel='Mes', xticks = np.arange(11) + 1,
       xticklabels = months, ylabel='Casos confirmados')

plt.show()
```

<div class="alert alert-success">
    
2. Hacer una grafica de barras comparando  los meses de agosto, septiembre, octubre y noviembre, considerando solo las entidades de Yucatán, Distrito Federal  y Jalisco
    
</div> 

```python
def data(df, estado):
    return df[estado].groupby(df.index.month).sum().loc[8:].values
```

```python
data(df, 'YUCATAN')
```

```python
# Datos para configuración
labels = ['Agosto', 'Septiembre','Octubre', 'Noviembre']
x_vals = np.arange(len(labels))  # Referencias para ubicación 
ancho = 0.3  # Ancho de las barras
```

```python
#Estilo Matlab
plt.figure(figsize=(8,5))  

plt.bar(x_vals - ancho, data(df, 'YUCATAN'), color = 'b', width = ancho,  label='YUCATÁN')
plt.bar(x_vals, data(df, 'DISTRITO FEDERAL'), color = 'g', width = ancho,  label='DISTRITO FEDERAL')
plt.bar(x_vals + ancho, data(df, 'JALISCO'), color = 'r', width = ancho,  label='JALISCO')

plt.xticks( x_vals , labels )
plt.xlabel('Mes')
plt.ylabel('Pacientes confirmados')
plt.legend()

#Titulo de la figura
plt.title('Casos confirmados de covid, por mes')

plt.show()
```

```python
#Estilo Orientado a Objetos
fig, ax = plt.subplots(figsize=(8,5), dpi=100)

ax.bar(x_vals - ancho, data(df, 'YUCATAN'), color = 'b', width = ancho,  label='YUCATÁN')
ax.bar(x_vals, data(df, 'DISTRITO FEDERAL'), color = 'g', width = ancho,  label='DISTRITO FEDERAL')
ax.bar(x_vals + ancho, data(df, 'JALISCO'), color = 'r', width = ancho,  label='JALISCO')

ax.set_xticks(x_vals)
ax.set_xticklabels(labels)
ax.legend()
ax.set(title='Casos confirmados de covid, por mes', xlabel= 'Mes' , ylabel='Pacientes confirmados')

plt.show()
```
