Muestre los 10 colegios con más aportaciones (o que más veces aparece)
    
</div>

```python
print('Los 10 colegios con más aportaciones son:')
for colegio in df.groupby('College').count().sort_values('Name', ascending=False).head(10).index:
    print('-', colegio)
```

<b>EJERCICIO 6 </b>: 
    
Agregar dos nuevas columnas:
    
a) Una nueva columna que se llame **Height(m)** y que tenga el valor en metros. La columna Height tiene unidades Pies-Pulgadas.
    
b) Una nueva columna que se llame **Weight(kg)** y que tenga el valor en Kilogramos. La columna Weight tiene unidades en libras.
</div>



```python
def ftin_to_m(height):
    val = height.split('-')
    return 0.3048*float(val[0]) + 0.0254*float(val[1])
```

```python
df['Height(m)'] = df['Height'].apply(ftin_to_m)
df
```

```python
df['Weight(kg)'] = df['Weight'] * 0.453592
df
```
