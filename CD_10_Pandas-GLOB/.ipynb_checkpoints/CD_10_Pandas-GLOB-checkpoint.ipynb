```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

## Mostrando todos los archivos de "/datos"

La salida serán los archivos en forma de lista

```python
# establecer ruta de búsqueda  
path = 'datos'
files = glob.glob(path + '/*')
files
```

## Mostrando  los archivos  **.txt** de "/datos"

La salida serán los archivos .txt en forma de lista

```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos txt  
path = 'datos'
files = glob.glob(path + '/*.txt')
files
```

## Mostrando  los archivos .csv de "/datos"

La salida serán los archivos .csv en forma de lista

```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files = glob.glob(path + '/*.csv')
files
```

## Concatenando todos los archivos .csv para formar un dataframe


```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files =  sorted(glob.glob(path + '/*.csv'))
print(files)

# crear una lista vacía para almacenar  los dataframes
lista_dfs = []
# recorrer la lista de archivos para crear la lista de archivos 
for file in files:
    temp_df = pd.read_csv(file)
    print(temp_df.shape)
    lista_dfs.append(temp_df)

# concatenación
df = pd.concat(lista_dfs, axis = 0)
print('Final:', df.shape)
df
```

## Concatenando los archivos .csv del primer semestre 

El comodín * representa desde nada, hasta cualquier cantidad de caracteres y dígitos

```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files = sorted(glob.glob(path + '/semestre1?.csv'))
print(files)
# crear una lista vacía para almacenar  los dataframes
lista_dfs = []
# recorrer la lista de archivos para crear la lista de archivos 
for file in files:
    temp_df = pd.read_csv(file)
    print(temp_df.shape)
    lista_dfs.append(temp_df)

# concatenación
df = pd.concat(lista_dfs, axis = 0)
print('Final:', df.shape)
df
```

## Concatenando los archivos .csv del primer y segundo semestre 

El comodín **[ ]**  representa un rango,  ya sea de caracteres o de números. 

```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files = sorted(glob.glob(path + '/semestre[1-2]?.csv'))
print(files)
# crear una lista vacía para almacenar  los dataframes
lista_dfs = []
# recorrer la lista de archivos para crear la lista de archivos 
for file in files:
    temp_df = pd.read_csv(file)
    print(temp_df.shape)
    lista_dfs.append(temp_df)

# concatenación
df = pd.concat(lista_dfs, axis = 0)
print('Final:', df.shape)
df
```

## Concatenando los archivos .csv de los grupos "a"
El comodín **?** Representa un único carácter.

```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files = sorted(glob.glob(path + '/semestre?a.csv'))
print(files)
# crear una lista vacía para almacenar  los dataframes
lista_dfs = []
# recorrer la lista de archivos para crear la lista de archivos 
for file in files:
    temp_df = pd.read_csv(file)
    print(temp_df.shape)
    lista_dfs.append(temp_df)

# concatenación
df = pd.concat(lista_dfs, axis = 0)
print('Final:', df.shape)
df
```

El comodín **[ ]**  representa un rango,  ya sea de caracteres o de números. 

```python
# establecer ruta de búsqueda  
# aquí queremos buscar archivos csv  
path = 'datos'
files = sorted(glob.glob(path + '/semestre?[a-b].csv'))
print(files)
# crear una lista vacía para almacenar  los dataframes
lista_dfs = []
# recorrer la lista de archivos para crear la lista de archivos 
for file in files:
    temp_df = pd.read_csv(file)
    print(temp_df.shape)
    lista_dfs.append(temp_df)

# concatenación
df = pd.concat(lista_dfs, axis = 0)
print('Final:', df.shape)
df
```

<div class="alert alert-success">
<b>EJERCICIO 1: Concatenar todos los archivos  .csv  </b>:

 <ul>
  <li> a) Encontrar el promedio y   desviación estándar  </li>
  <li> b) Datos del estudiante(s) con la calificación más alta </li>   
  <li> c) Datos del estudiante(s) con la calificación más baja </li>  
  <li> d) El nombre que más se repite </li>  
  <li> e) El apellido que más se repite </li>       
     
</ul>
</div>

```python
# Encontrar los nombres de los archivos
files = sorted(glob.glob('datos/semestre??.csv'))
print(files)

# Crear una lista con los DataFrames
lista_dfs = []
for file in files:
    temp_df = pd.read_csv(file)
    lista_dfs.append(temp_df)
    
# Creando el DataFrame
df = pd.concat(lista_dfs, axis = 0)
df.head()
```

```python
print('El promedio de calificaciones es:', df['CALIFICACION'].mean())
print('La desviación estandar de las calificaciones es:', df['CALIFICACION'].std())
```

```python
print('Los estudiantes con la calificación más alta son:')
df[df['CALIFICACION'].max() == df['CALIFICACION']]
```

```python
print('Y con la calificación más baja:')
df[df['CALIFICACION'].min() == df['CALIFICACION']]
```

```python
print('Los nombres que más se repiten son:')
for name in df['NOMBRE'].value_counts()[df['NOMBRE'].value_counts() == df['NOMBRE'].value_counts()[0]].index:
    print('-', name)
```

```python
print('Los apellidos que más se repiten son:')
for apellido in df['APELLIDO'].value_counts()[df['APELLIDO'].value_counts() == df['APELLIDO'].value_counts()[0]].index:
    print('-', apellido)
```
